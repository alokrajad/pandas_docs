```python
import pandas as pd
```

```python
dict2=[{"a":1,"b":2},{"a":4,"d":5,"c":5},{"a":4,"d":8,"c":1},{"a":2,"b":8,"c":4},{"a":6,"b":7,"c":2}]
df1=pd.DataFrame(dict2)
df1
```

```python
df1.fillna(0)     # it wll chnge all nan to 0. values=scalar, dict, Series, or DataFrame.
```

```python
df1.fillna({'b':'None',"c":'null'}) 
```

```python
df1
```

```python
#method : {'backfill', 'bfill', 'pad', 'ffill', None}, default None
# pad / ffill: propagate last valid observation forward to next valid
# backfill / bfill: use NEXT valid observation to fill gap


df1.fillna(method='ffill') 
```

```python
df1.fillna(method='pad') 
```

```python
df1.fillna(method='backfill') 
```

```python
df1.fillna(method='bfill') 
```

```python
 df1.fillna(axis=0) #error:Must specify a fill 'value' or 'method'. axis will work with value or method
```

```python
df1.fillna(value=0,axis=1)
```

```python
df1
```

```python
df1.fillna(method='ffill',axis=1) 
```

```python
#limit=n  it filled only the n number of nan.it works with value and method 
df1.fillna(5,limit=1) 
```

```python
df1
```

```python
df1.fillna(4,limit=1)   #it fill every nan in each column to the given number of limit
```
