```python
dict2=[{"a":1,"b":2},{"a":4,"d":5,"c":5},{"a":4,"d":8,"c":1},{"a":2,"b":8,"c":4},{"a":6,"b":7,"c":2}]
df1=pd.DataFrame(dict2)
df1
```

```python
df1.dropna()
```

```python
df1.dropna(axis=0,how='any')   
```

```python
#axis=0 and how='any' both are by default.thts why both upper results are same
```

```python
df1.dropna(axis=1,how='any')
```

```python
df1.dropna(how='all')    #how=all will drop that row or column which have all value Nan
```

```python
#subset
```

```python
df1.dropna(subset=['c','d'])  #subset take column name in a list and drop all column which have Nan
```

```python
#thresh
```

```python
df1.dropna(thresh=3)  #atleast 3 non nan value will be there. 
```

```python
df1
```

```python
df1.dropna(axis=1,inplace=True) #inplcae=True chnage the orignal data frame
```
