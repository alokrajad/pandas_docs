```python
import pandas as pd
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],'Name':["A","B","C","D"]})
df1
```

```python
df2=pd.DataFrame({'ID':[1,2,3,4],'AGE':[22,23,21,32]})
df2
```

```python
df1.set_index('ID').join(df2.set_index('ID'))
```

```python
df1.join(df2,lsuffix='_name', rsuffix='_age')
```

```python
df1.join(df2.set_index('ID'), on='ID')
```

```python
df1.join(df2,how='left',lsuffix='_name', rsuffix='_age')
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],'Name':["A","B","C","D"]})
```

```python
df2=pd.DataFrame({'ID':[1,2,3],'AGE':[22,23,21]})
```

```python
df1.join(df2,how='left',lsuffix='_name', rsuffix='_age')
```

```python
df1.join(df2,how='right',lsuffix='_name', rsuffix='_age')
```

```python
df1.join(df2,how='outer',lsuffix='_name', rsuffix='_age')
```

```python
df1.join(df2,how='inner',lsuffix='_name', rsuffix='_age')
```

```python
df = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                  'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
df
```

```python
other = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                  'B': ['B0', 'B1', 'B2']})
other
```

```python
df.join(other, lsuffix='_caller', rsuffix='_other')
```

```python
df.set_index('key').join(other.set_index('key'))
```
