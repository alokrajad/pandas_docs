```python
 dict2=[{"a":"hii","b":2},{"a":4,"d":5,"c":5},{"a":4,"d":8,"c":1},{"a":2,"b":8,"c":4},{"a":6,"b":7,"c":2}]
df=pd.DataFrame(dict2)
df
```

```python
 dict2=[{"a":"hii","b":2},{"a":"hj","d":5,"c":"ghjkjk"},{"a":4,"d":"dfd","c":1},{"a":2,"b":"hjgj","c":4},{"a":6,"b":7,"c":2}]
df=pd.DataFrame(dict2)
df
```

```python
df.replace("[A-Za-z]",0) #trying to chnge all string by using pattern bt it dint chnged here.
```

```python
df.replace("[A-Za-z]",0,regex=True) #it take regularexpresion True parmeter.now it chnged
```

```python
df.replace({'a':'[A-Za-z]'},0,regex=True) #specific column string chnge using dict
```

```python
df.replace('hii',method='backfill')
```

```python
df.replace(2,method='ffill',limit=2)  #limit=2 means only two value will chnge
```
