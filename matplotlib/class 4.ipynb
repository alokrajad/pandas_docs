```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
sub=['python','R','AI','ML','DS']
class1=[35,45,36,56,52]
class2=[65,45,65,37,68]
class3=[56,76,45,65,67]
```

```python
plt.bar(sub,class1)
plt.show()
```

```python
#horizontol
plt.barh(sub,class1)
plt.show()
```

```python
#width
plt.bar(sub,class1,width=0.2)
plt.show()
```

```python
#align(by default'centre','edge')
plt.bar(sub,class1,width=0.2,align='edge')
plt.show()
```

```python
#color
plt.bar(sub,class1,width=0.2,align='edge',color='y')
plt.show()
```

```python
#edgecolor
plt.bar(sub,class1,width=0.2,align='edge',color='y',edgecolor='m')
plt.show()
```

```python
#linewidth
plt.bar(sub,class1,width=0.2,align='edge',color='y',edgecolor='m',linewidth=5)
plt.show()
```

```python
#alpha (color of graph b/w 0-1)
plt.bar(sub,class1,width=0.2,align='edge',color='y',edgecolor='m',linewidth=5,alpha=0.2)
plt.show()
```

```python
#linestyle
plt.bar(sub,class1,width=0.2,align='edge',color='y',edgecolor='m',
        linewidth=5,alpha=0.9,linestyle='--')
plt.show()
```

```python
#label
plt.bar(sub,class1,width=0.2,align='edge',color='y',edgecolor='m',linewidth=5,alpha=0.9,label='class 1')
plt.legend()
plt.show()
```

```python
#visible
plt.bar(sub,class1,width=0.2,
        align='edge',color='y',edgecolor='m',linewidth=5,
        alpha=0.9,label='class 1',visible=False)
plt.show()
```

```python
#background
style.use('ggplot')
plt.figure(figsize=(16,9))
plt.bar(sub,class1,width=0.2,
        align='edge',color='k',edgecolor='m',linewidth=16,
        alpha=0.9,label='class 1',visible=True)
plt.show()
```

```python
#horizontal
style.use('ggplot')
plt.figure(figsize=(16,9))
plt.barh(sub,class1,
        align='edge',color='k',edgecolor='m',linewidth=5,
        alpha=0.9,label='class 1',visible=True)#remove width else get eror
plt.show()
```

```python
#label
style.use('ggplot')
plt.figure(figsize=(16,9))
plt.bar(sub,class1,width=0.2,
        align='edge',color='k',edgecolor='m',linewidth=5,
        alpha=0.9,label='class 1',visible=True)
plt.title("student details",fontsize=28)
plt.xlabel("student details",fontsize=18)
plt.ylabel("student details",fontsize=18)
plt.show()
```

```python
#multiple bar graph
plt.bar(sub,class1,width=0.2,
        align='edge',color='g',edgecolor='m',linewidth=5,
        alpha=0.9,label='class 1',visible=True)
plt.bar(sub,class2,width=0.2,
        align='edge',color='r',edgecolor='m',linewidth=5,
        alpha=0.9,label='class 1',visible=True)
plt.title("student details",fontsize=28)
plt.xlabel("student details",fontsize=18)
plt.ylabel("student details",fontsize=18)
plt.show()
```

```python
#multiple bar graph
plt.figure(figsize=(16,9))
sub_index=np.arange(len(sub))
width=0.2
plt.bar(sub_index,class1,width,
        align='edge',color='g',edgecolor='m',label='class 1',visible=True)
plt.bar(sub_index+width,class2,width,
        align='edge',color='r',edgecolor='m',label='class 1',visible=True)
plt.title("student details",fontsize=28)
plt.xlabel("student details",fontsize=18)
plt.ylabel("student details",fontsize=18)
plt.show()
```

```python
#replace class index by name &rotate
plt.figure(figsize=(16,9))
sub_index=np.arange(len(sub))
width=0.2
plt.bar(sub_index,class1,width,
        align='edge',color='g',edgecolor='m',label='class 1',visible=True)
plt.bar(sub_index+width,class2,width,
        align='edge',color='r',edgecolor='m',label='class 1',visible=True)
plt.bar(sub_index+width+width,class3,width,
        align='edge',color='b',edgecolor='b',label='class 1',visible=True)
plt.xticks(sub_index+width,sub,rotation=20)
plt.title("student details",fontsize=28)
plt.xlabel("student details",fontsize=18)
plt.ylabel("student details",fontsize=18)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sub_index=np.arange(len(sub))
width=0.2
plt.barh(sub_index,class1,width,
        align='edge',color='g',edgecolor='m',label='class 1',visible=True)
plt.barh(sub_index+width,class2,width,
        align='edge',color='r',edgecolor='m',label='class 2',visible=True)
plt.barh(sub_index+width+width,class3,width,
        align='edge',color='b',edgecolor='b',label='class 3',visible=True)
plt.yticks(sub_index+width,sub,rotation=20)
plt.title("student details",fontsize=28)
plt.ylabel("student details",fontsize=18)
plt.xlabel("student details",fontsize=18)
plt.legend()
plt.show()
```
