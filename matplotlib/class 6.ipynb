```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import style
```

```python
plt.pie([10])
plt.show()
```

```python
sub=['PY','R','ML','AI',"DS"]
std=[45,15,35,25,30]
```

```python
plt.pie(std)
plt.show()
```

```python
#labels
sub=['PY','R','ML','AI',"DS"]
std=[45,15,35,25,30]
plt.pie(std,labels=sub)
plt.show()
```

```python
#explode
sub=['PY','R','ML','AI',"DS"]
std=[45,15,35,25,30]
x=[0,0,0.1,0,0]
plt.pie(std,labels=sub,explode=x)
plt.show()
```

```python
#colors
sub=['PY','R','ML','AI',"DS"]
std=[45,15,35,25,30]
x=[0.1,0,0.2,0,0]
colors=['c','b','r','y','g']
plt.pie(std,labels=sub,explode=x,colors=colors)
plt.show()
```

```python
#auto percentage
sub=['PY','R','ML','AI',"DS"]
std=[45,15,35,25,30]
x=[0.1,0,0.2,0,0]
colors=['c','b','r','y','g']
plt.pie(std,labels=sub,explode=x,colors=colors,autopct='%0.1f%%')
plt.show()
```

```python
#shadow
sub=['PY','R','ML','AI',"DS"]
std=[45,15,35,25,30]
x=[0.1,0,0.2,0,0]
colors=['c','b','r','y','g']
plt.pie(std,labels=sub,explode=x,colors=colors,autopct='%0.1f%%',shadow=True)
plt.show()
```

```python
#radius
sub=['PY','R','ML','AI',"DS"]
std=[45,15,35,25,30]
x=[0.1,0,0.2,0,0]
colors=['c','b','r','y','g']
plt.pie(std,labels=sub,explode=x,colors=colors,autopct='%0.1f%%',shadow=True,radius=1.5)
plt.show()
```

```python
#startangle)
sub=['PY','R','ML','AI',"DS"]
std=[45,15,35,25,30]
x=[0.1,0,0.2,0,0]
colors=['c','b','r','y','g']
plt.pie(std,labels=sub,explode=x,colors=colors,autopct='%0.1f%%',
        shadow=True,radius=1.5,startangle=90)
plt.show()
```

```python
#textprops
sub=['PY','R','ML','AI',"DS"]
std=[45,15,35,25,30]
x=[0.1,0,0.2,0,0]
colors=['c','b','r','y','g']
t={"fontsize":15}
plt.pie(std,labels=sub,explode=x,colors=colors,autopct='%0.1f%%',
        shadow=True,radius=1.5,startangle=90,textprops=t)
plt.show()
```

```python
#pctdistance
sub=['PY','R','ML','AI',"DS"]
std=[45,15,35,25,30]
x=[0.1,0,0.2,0,0]
colors=['c','b','r','y','g']
t={"fontsize":15}
plt.pie(std,labels=sub,explode=x,colors=colors,autopct='%0.1f%%',
        shadow=True,radius=1.5,startangle=90,textprops=t,pctdistance=0.8)
plt.show()
```

```python
#labeldistance
sub=['PY','R','ML','AI',"DS"]
std=[45,15,35,25,30]
x=[0.1,0,0.2,0,0]
colors=['c','b','r','y','g']
t={"fontsize":15}
plt.pie(std,labels=sub,explode=x,colors=colors,autopct='%0.1f%%',
        shadow=True,radius=1.5,startangle=90,textprops=t,pctdistance=0.8,labeldistance=1.2)
plt.show()
```

```python
#wedgeprops
wedgeprops={"linewidth":5,"width":3,"edgecolor":"k"}
sub=['PY','R','ML','AI',"DS"]
std=[45,15,35,25,30]
x=[0.1,0,0.2,0,0]
colors=['c','b','r','y','g']
t={"fontsize":15}
plt.pie(std,labels=sub,explode=x,
        colors=colors,
        autopct='%0.1f%%', 
    shadow=True,radius=1.5,startangle=90,counterclock=True,wedgeprops=wedgeprops,
        textprops=t,pctdistance=0.8,
        labeldistance=1.2)
plt.show()
```

```python
plt.figure(figsize=(8,4))
wedgeprops={"linewidth":5,"width":3,"edgecolor":"k"}
sub=['PY','R','ML','AI',"DS"]
std=[45,15,35,25,30]
x=[0.1,0,0.2,0,0]
colors=['c','b','r','y','g']
t={"fontsize":15}
plt.pie(std,labels=sub,explode=x,
        colors=colors,
        autopct='%0.1f%%', 
    shadow=True,radius=1.5,startangle=90,counterclock=True,wedgeprops=wedgeprops,
        textprops=t,pctdistance=0.8,
        labeldistance=1.2)
plt.show()
```

```python
sub=['PY','R','ML','AI',"DS"]
std=[45,15,35,25,30]
x=[0.1,0,0.2,0,0]
colors=['c','b','r','y','g']
t={"fontsize":15}
plt.pie(std,labels=sub,explode=x,colors=colors,autopct='%0.1f%%',
        shadow=True,radius=1.5,startangle=90,textprops=t,pctdistance=0.8,labeldistance=1.2)
plt.legend()
plt.show()
```
