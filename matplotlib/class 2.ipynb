```python
#color
days=[1,2,3,4,5]
temp=[65.5,38.6,45.6,47.6,76.5]
plt.axis([0,6,0,80])
plt.title("Delhi temprature")
plt.xlabel("days")
plt.ylabel("temp")
plt.plot(days,temp,color='r')
plt.show()
```

```python
#marker
days=[1,2,3,4,5]
temp=[65.5,38.6,45.6,47.6,76.5]
plt.axis([0,6,0,80])
plt.title("Delhi temprature")
plt.xlabel("days")
plt.ylabel("temp")
plt.plot(days,temp,color='r',marker='D')
plt.show()
```

```python
#linestyle
days=[1,2,3,4,5]
temp=[65.5,38.6,45.6,47.6,76.5]
plt.axis([0,6,0,80])
plt.title("Delhi temprature")
plt.xlabel("days")
plt.ylabel("temp")
plt.plot(days,temp,color='r',marker='D',linestyle=':')
plt.show()
```

```python
#linewidth
days=[1,2,3,4,5]
temp=[65.5,38.6,45.6,47.6,76.5]
plt.axis([0,6,0,80])
plt.title("Delhi temprature")
plt.xlabel("days")
plt.ylabel("temp")
plt.plot(days,temp,color='r',marker='D',linestyle=':',linewidth=4)
plt.show()
```

```python
#markersize
days=[1,2,3,4,5]
temp=[65.5,38.6,45.6,47.6,76.5]
plt.axis([0,6,0,80])
plt.title("Delhi temprature")
plt.xlabel("days")
plt.ylabel("temp")
plt.plot(days,temp,color='r',marker='D',linestyle=':',linewidth=4,markersize=10)
plt.show()
```

```python
#shorcut
days=[1,2,3,4,5]
temp=[65.5,38.6,45.6,47.6,76.5]
plt.axis([0,6,0,80])
plt.title("Delhi temprature")
plt.xlabel("days")
plt.ylabel("temp")
plt.plot(days,temp,'rD:',linewidth=4)
plt.show()
```

```python
#titlesize
days=[1,2,3,4,5]
temp=[65.5,38.6,45.6,47.6,76.5]
plt.axis([0,6,0,80])
plt.title("Delhi temprature",fontsize=15)
plt.xlabel("days",fontsize=15)
plt.ylabel("temp",fontsize=15)
plt.plot(days,temp,'rD:',linewidth=4)
plt.show()
```

```python
#legend(line name) 
days=[1,2,3,4,5]
temp=[65.5,38.6,45.6,47.6,76.5]
plt.axis([0,6,0,80])
plt.title("Delhi temprature",fontsize=15)
plt.xlabel("days",fontsize=15)
plt.ylabel("temp",fontsize=15)
plt.plot(days,temp,'D:',linewidth=4)#bydefault color 
plt.legend(["Temp line"])
plt.show()
```

```python
#legend location
days=[1,2,3,4,5]
temp=[65.5,38.6,45.6,47.6,76.5]
plt.axis([0,6,0,80])
plt.title("Delhi temprature",fontsize=15)
plt.xlabel("days",fontsize=15)
plt.ylabel("temp",fontsize=15)
plt.plot(days,temp,'D:',linewidth=4)
plt.legend(["Temp line"],loc=0)#location value is in documentation 0-10
plt.show()
```

```python
#bg color
from matplotlib.pyplot import style
days=[1,2,3,4,5]
temp=[65.5,38.6,45.6,47.6,76.5]
style.use("ggplot")
plt.axis([0,6,0,80])
plt.title("Delhi temprature",fontsize=15)
plt.xlabel("days",fontsize=15)
plt.ylabel("temp",fontsize=15)
plt.plot(days,temp,'gD:',linewidth=4) 

plt.show()
```

```python
#grid color
from matplotlib.pyplot import style
days=[1,2,3,4,5]
temp=[65.5,38.6,45.6,47.6,76.5]
style.use("ggplot")
plt.axis([0,6,0,80])
plt.title("Delhi temprature",fontsize=15)
plt.xlabel("days",fontsize=15)
plt.ylabel("temp",fontsize=15)
plt.plot(days,temp,'gD:',linewidth=4,label="temp") 
plt.grid(color='r', linestyle='-', linewidth=2)
plt.legend(loc=3)
plt.show()
```

```python
#double grpah
from matplotlib.pyplot import style
days=[1,2,3,4,5]
temp=[65.5,38.6,45.6,47.6,76.5]
humid=[23,45,67,56,53]
style.use("classic")
plt.axis([0,6,0,80])
plt.title("Delhi temprature & humid",fontsize=15)
plt.xlabel("days",fontsize=15)
plt.ylabel("temp Y humid",fontsize=15)
plt.plot(days,temp,'bD-',label="templine") 
plt.plot(days,humid,'gD-',linewidth=4,label='hline')
plt.grid(color='r', linestyle='-', linewidth=2)
plt.legend(loc=3)
plt.show()
```

```python
#double grpah
days=[1,2,3,4,5]
temp=[65.5,38.6,45.6,47.6,76.5]
humid=[23,45,67,56,53]
plt.axis([0,6,0,80])
plt.title("Delhi temprature & humid",fontsize=15)
plt.xlabel("days",fontsize=15)
plt.ylabel("temp Y humid",fontsize=15)
plt.plot(days,temp,'bD-', label='line 1',linewidth=4) 
plt.plot(days,humid,'gD-',linewidth=4)


plt.show()
```
