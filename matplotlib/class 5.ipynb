```python
#scatter plot
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import style
```

```python
df=pd.read_csv('C:\\Users\\acer\\Desktop\\python\\pandas\\data.csv')
df.shape
```

```python
 df.head()
```

```python
x=df['Maxpulse']
y=df['Pulse']
plt.scatter(x,y)
plt.title("pulse Data")
plt.ylabel("pulse")
plt.xlabel("maxpulse")
plt.show()
```

```python
#color
x=df['Maxpulse']
y=df['Pulse']
plt.scatter(x,y,c='r')
plt.title("pulse Data")
plt.ylabel("pulse")
plt.xlabel("maxpulse")
plt.show()
```

```python
#scatter plot size
plt.figure(figsize=(16,9))
x=df['Maxpulse']
y=df['Pulse']
plt.scatter(x,y,c='r')
plt.title("pulse Data")
plt.ylabel("pulse")
plt.xlabel("maxpulse")
plt.show()
```

```python
#marker size
plt.figure(figsize=(16,9))
x=df['Maxpulse']
y=df['Pulse']
plt.scatter(x,y,c='r',marker="*",s=500)
plt.title("pulse Data")
plt.ylabel("pulse")
plt.xlabel("maxpulse")
plt.show()
```

```python
#transparancy of marker
plt.figure(figsize=(16,9))
x=df['Maxpulse']
y=df['Pulse']
plt.scatter(x,y,c='r',marker="*",s=200,alpha=0.5)
plt.title("pulse Data")
plt.ylabel("pulse")
plt.xlabel("maxpulse")
plt.show()
```

```python
#linewidth of marker 
plt.figure(figsize=(16,9))
x=df['Maxpulse']
y=df['Pulse']
plt.scatter(x,y,c='r',marker="*",s=200,alpha=0.5,linewidth=11)
plt.title("pulse Data")
plt.ylabel("pulse")
plt.xlabel("maxpulse")
plt.show()
```

```python
#edge color
plt.figure(figsize=(16,9))
x=df['Maxpulse']
y=df['Pulse']
plt.scatter(x,y,c='r',marker="*",s=200,edgecolor='g')
plt.title("pulse Data")
plt.ylabel("pulse")
plt.xlabel("maxpulse")
plt.show()
```

```python
plt.figure(figsize=(16,9))
x=df['Maxpulse']
y=df['Pulse']
z=df['Duration']
plt.scatter(x,y,c='r',marker="*",s=200,edgecolor='g')
plt.scatter(x,z,c='r',marker="*",s=200,edgecolor='r')
plt.title("pulse Data")
plt.ylabel("pulse")
plt.xlabel("maxpulse")
plt.show()
```
