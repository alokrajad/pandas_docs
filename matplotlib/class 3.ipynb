```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
s1_student_age=np.random.randint(18,45,(100))
s2_student_age=np.random.randint(15,40,(100))

print(s1_student_age)
print(s2_student_age)
```

```python
plt.hist(s1_student_age)

plt.title("Student Age")
plt.xlabel("student age category")
plt.ylabel("No. student ")
plt.show()
```

```python
#bins--> x axis width
bins=[15,20,25,30,40,45,50]
plt.hist(s1_student_age,bins,histtype='bar')
plt.title("Student Age")
plt.xlabel("student age category")
plt.ylabel("No. student ")

plt.show()
```

```python
#rwidth
bins=[15,20,25,30,35,40,45,50]
plt.hist(s1_student_age,bins,rwidth=0.9)
plt.title("Student Age")
plt.xlabel("student age category")
plt.ylabel("No. student ")

plt.show()
```

```python
#histtype--> type of histogram(bar,barstacked,step,stepfilled)
plt.hist(s1_student_age,bins,histtype="step",rwidth=0.8)
plt.title("Student Age")
plt.xlabel("student age category")
plt.ylabel("No. student ")

plt.show()
```

```python
#orientation-->horizonatol graph' 
plt.hist(s1_student_age,bins,histtype="bar",rwidth=0.8,orientation='horizontal')
plt.title("Student Age")
plt.xlabel("student age category")
plt.ylabel("No. student ")

plt.show()
```

```python
#color & label
plt.hist(s1_student_age,bins,histtype="bar",rwidth=0.8,orientation='horizontal',color='m',label='student age')
plt.title("Student Age")
plt.xlabel("student age category")
plt.ylabel("No. student ")
plt.legend()
plt.show()
```

```python
#doublebar graph
plt.hist(s1_student_age,bins,histtype="bar",rwidth=0.8,orientation='vertical',color='m')
plt.hist(s2_student_age,bins,histtype="bar",rwidth=0.8,orientation='vertical',color='r')

plt.title("Student Age")
plt.xlabel("student age category")
plt.ylabel("No. student ")

plt.show()
```

```python
#doublebar graph
plt.hist([s1_student_age,s2_student_age],bins,histtype="bar",orientation='vertical',color=['m','r'])
# plt.hist(s2_student_age,bins,histtype="bar",rwidth=0.8,orientation='vertical',color='r')

plt.title("Student Age")
plt.xlabel("student age category")
plt.ylabel("No. student ")

plt.show()
```

```python
#size increase
plt.figure(figsize=(16,9))
plt.hist([s1_student_age,s2_student_age],bins,histtype="bar",
         rwidth=0.8,orientation='vertical',color=['m','r'],label=['s1 student age','s2 student age'])
# plt.hist(s2_student_age,bins,histtype="bar",rwidth=0.8,orientation='vertical',color='r')

plt.title("Student Age")
plt.xlabel("student age category")
plt.ylabel("No. student ")
plt.legend()
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
plt.figure(figsize=(16,9))
plt.hist([s1_student_age,s2_student_age],bins,histtype="bar",rwidth=0.8,orientation='vertical',color=['m','r'],label=['s1 student age','s2 student age'])
# plt.hist(s2_student_age,bins,histtype="bar",rwidth=0.8,orientation='vertical',color='r')

plt.title("Student Age")
plt.xlabel("student age category")
plt.ylabel("No. student ")
plt.grid(color='b', linestyle='-', linewidth=2)
plt.legend()
plt.show()
```
